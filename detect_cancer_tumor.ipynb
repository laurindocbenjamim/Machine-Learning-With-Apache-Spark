```python
# You can also use this section to suppress warnings generated by your code:
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
warnings.filterwarnings('ignore')

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
# the data set is available at the url below.
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-BD0231EN-SkillsNetwork/datasets/cancer.csv"

# using the read_csv function in the pandas library, we load the data into a dataframe.

df = pd.read_csv(URL)
```

```python
# show 5 random rows from the dataset
df.sample(5)
```

```python
# Find  out the number of  rows and columns
df.shape
```

```python
target = df["diagnosis"]
```

```python
# We identify the features next. Features are the input values our machine learning model learns from
features = df[['radius_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'symmetry_mean']]
```

```python
# Create a Logistic Regression model
classifier = LogisticRegression()
```

```python
# Train/Fit the model
classifier.fit(features, target)
```

```python
# Evaluate the model
#Higher the score, better the model.
classifier.score(features,target)
```

Make predictions. Let us predict the species of a flower with SepalLengthCm = 5.4, SepalWidthCm = 2.6, PetalLengthCm = 4.1, PetalWidthCm = 1.3

```python
classifier.predict([[13.45, 86.6, 555.1, 0.1022, 0.08165, 0.03974, 0.1638]])
```
